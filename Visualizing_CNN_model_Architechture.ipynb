```python
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import plot_model

# Load the saved model
model = load_model("/content/dual_input_brick_classifier.h5")

# Plot the architecture
plot_model(model,
           to_file="dual_input_2dcnn_architecture.png",
           show_shapes=True,
           show_layer_names=True,
           expand_nested=True,
           dpi=100)
```
